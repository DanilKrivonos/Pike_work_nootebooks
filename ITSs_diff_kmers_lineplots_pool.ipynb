```python
def get_color(obj_dict):
    color = ''
    while color not in obj_dict.values() and color == '':
        color = "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
    return color
color_dict_its1_s = {}
color_dict_its1_otu = {}
color_dict_its2_s = {}
color_dict_its2_otu = {}
color_dict_its12_s = {}
color_dict_its12_otu = {}
```

```python
for taxon in mean_reampl_free_its1_s.index:
    color_dict_its1_s[taxon] = get_color(color_dict_its1_s)
for taxon in mean_reampl_free_its1_otu.index:
    color_dict_its1_otu[taxon] = get_color(color_dict_its1_otu)
for taxon in mean_reampl_free_its2_s.index:
    color_dict_its2_s[taxon] = get_color(color_dict_its2_s)
for taxon in mean_reampl_free_its2_otu.index:
    color_dict_its2_otu[taxon] = get_color(color_dict_its2_otu)
for taxon in mean_reampl_free_its12_s.index:
    color_dict_its12_s[taxon] = get_color(color_dict_its12_s)
for taxon in mean_reampl_free_its12_otu.index:
    color_dict_its12_otu[taxon] = get_color(color_dict_its12_otu)
```
