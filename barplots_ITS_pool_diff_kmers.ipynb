```python
def extract_key(column_name):
    match = re.match(r'(.+?)(?:_(\d+(?:\.\d+)?))?_reads', column_name)
    if match:
        sample = match.group(1)
        n = int(match.group(2))
        return sample, n
    return column_name, float('inf')
```
