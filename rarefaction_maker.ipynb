**MAKING LONG DFs FOR RAREFACTION PLOTS, POOL MODE**

```python
import pandas as pd
import numpy as np
```

```python
# read data_tax_dfs
data_tax_df = pd.read_csv('/home/sonec_iv/pike_ipynbs/pool_ITS_1_data_tax_df_genus_new_db.csv',sep ='\t',index_col=0)
data_tax_df
```

```python
#levels
#level = 'Genus'
#evel = 'Species'
level = 'OTU'
```

```python
postfix = 'rarefactn'
#prefix  = 'single'
prefix = 'pool'
```

```python
#sample = 'V1_V9'
#sample = 'V3_V4'
sample = 'ITS1'
#sample = 'ITS2'
#sample = 'ITS1_5.8_ITS2'
```

```python
data_tax_no0_rows = (data_tax_df !=0).sum()
result = pd.DataFrame(data_tax_no0_rows).transpose()
result_long = result.melt(var_name='column', value_name=f'{level}_count_nonzero')
result_long['sample_name'] = result_long['column'].str.extract(rf'({sample}_(?:R)?\d+)_\d+_reads')
result_long['number_of_reads'] = result_long['column'].str.extract(rf'{sample}_(?:R)?\d+_(\d+)_reads')
result_long.drop('column', axis=1, inplace=True)
result_long.fillna(0)
result_long['number_of_reads'] = result_long['number_of_reads'].astype('int')
result_long=result_long.sort_values('number_of_reads',ascending=True)
result_long.head(10)
```

```python
result_long.to_csv(f'{prefix}_{sample}_{level}_{postfix}_count_long_table.csv', sep='\t',index=False)
```
