```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import umap
import hdbscan
import sklearn.cluster as cluster
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score
```

```python
import random
def get_color(obj_dict):
    
    color = ''
    
    while color not in obj_dict.values() and color == '':
        
        color = "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
    
    return color
```

```python
metadata_tsv_ITS1_10k_single = pd.read_csv('/mnt/AsusShareI2/RUNS/runs-sonec/pike_single_mode/pike_ITS1_all/pike_ITS1_trimmed_10000_reads/work_dir/ITS1_1/metadata_df.tsv', sep = '\t')
#metadata_tsv_ITS1_10k_single
metadata_tsv_ITS1_single_k3 = pd.read_csv('/mnt/AsusShareI2/RUNS/runs-sonec/pike_res_diff_kmers/ITS1/pike_ITS1_trimmed_3_kmer_size/work_dir/ITS1_1/metadata_df.tsv', sep = '\t')
#metadata_tsv_ITS1_10k_single_k3
metadata_tsv_ITS1_single_k6 = pd.read_csv('/mnt/AsusShareI2/RUNS/runs-sonec/pike_res_diff_kmers/ITS1/pike_ITS1_trimmed_6_kmer_size/work_dir/ITS1_1/metadata_df.tsv', sep = '\t')
#metadata_tsv_ITS1_10k_single_k6
```

```python
color_collection_its1_10k_clusters = {}
color_collection_its1_k3_clusters = {}
color_collection_its1_k6_clusters = {}
```

```python
for i in metadata_tsv_ITS1_10k_single['Clusters']:
    color_collection_its1_10k_clusters[i] = get_color(color_collection_its1_10k_clusters)
for i in metadata_tsv_ITS1_10k_single_k3['Clusters']:
    color_collection_its1_k3_clusters[i] = get_color(color_collection_its1_k3_clusters)
for i in metadata_tsv_ITS1_10k_single_k6['Clusters']:
    color_collection_its1_k6_clusters[i] = get_color(color_collection_its1_k6_clusters)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=metadata_tsv_ITS1_10k_single, x='1 UMAP COMPONENT', y='2 UMAP COMPONENT', palette=color_collection_its1_10k_clusters, hue = 'Clusters', s=15)
plt.title('ITS1 10000 reads UMAP clustering results')
plt.legend().remove()
plt.savefig("ITS1_UMAP_10k_reads_SINGLE.pdf",format="pdf", bbox_inches="tight")
plt.savefig("ITS1_UMAP_10k_reads_SINGLE.png", dpi=800, format="png", bbox_inches='tight')
```
