```python
df = pd.DataFrame(data_its1_g)
def calculate_means(df):
    result = {}

    # ÐŸÐ¾Ð»ÑƒÑ‡Ð°ÐµÐ¼ ÑƒÐ½Ð¸ÐºÐ°Ð»ÑŒÐ½Ñ‹Ðµ numbers Ð¸Ð· Ð½Ð°Ð·Ð²Ð°Ð½Ð¸Ð¹ ÑÑ‚Ð¾Ð»Ð±Ñ†Ð¾Ð²
    numbers = df.columns.str.split('_').str[2].drop_duplicates().dropna()

    for number in numbers:
        # Ð¡Ñ‚Ð¾Ð»Ð±Ñ†Ñ‹ Ð´Ð»Ñ Ñ‚Ð¸Ð¿Ð¾Ð² 1, 2, 3
        group_123 = pd.concat([
            df.filter(like=f'_{number}_').filter(like='_1_'),
            df.filter(like=f'_{number}_').filter(like='_2_'),
            df.filter(like=f'_{number}_').filter(like='_3_')
        ], axis=1)

        result[f'ITS1_{number}_mean'] = group_123.mean(axis=1)

        # Ð¡Ñ‚Ð¾Ð»Ð±Ñ†Ñ‹ Ð´Ð»Ñ Ñ‚Ð¸Ð¿Ð¾Ð² R1, R2, R3
        group_R = pd.concat([
            df.filter(like=f'_{number}_').filter(like='_R1_'),
            df.filter(like=f'_{number}_').filter(like='_R2_'),
            df.filter(like=f'_{number}_').filter(like='_R3_')
        ], axis=1)

        result[f'ITS1_{number}_R_mean'] = group_R.mean(axis=1)

    return pd.DataFrame(result)

# Ð’Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÐ¼ ÑÑ€ÐµÐ´Ð½Ð¸Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ
mean_df = calculate_means(data_its1_g)
mean_df = mean_df.reindex(sorted(mean_df.columns, key=lambda x: int(x.split('_')[1])), axis=1)
```
